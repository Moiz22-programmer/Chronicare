# ChroniCare: Supervised Model Training Pipeline
This notebook details the training pipelines for the four chronic disease classifiers of ChroniCare:
1. **Diabetes**: Random Forest (Primary) vs. Logistic Regression (Baseline)
2. **Heart Disease**: Support Vector Machine (Primary) vs. Decision Tree (Baseline)
3. **Kidney Disease (CKD)**: K-Nearest Neighbors (Primary) vs. Gaussian Naive Bayes (Baseline)
4. **Liver Disease**: XGBoost (Primary) vs. Gradient Boosting (Baseline)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import pickle
import os
```

## 1. Diabetes Prediction Model
We train a **Random Forest Classifier** and compare it with a **Logistic Regression** baseline.

```python
df_diab = pd.read_csv('../data/diabetes.csv')
print("Diabetes Dataset Shape:", df_diab.shape)
X = df_diab.drop(columns=['Outcome'])
y = df_diab['Outcome']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

rf = RandomForestClassifier(n_estimators=100, max_depth=8, random_state=42)
rf.fit(X_train, y_train)
print("Random Forest Train Accuracy:", rf.score(X_train, y_train))
print("Random Forest Test Accuracy:", rf.score(X_test, y_test))

lr = LogisticRegression(max_iter=1000, random_state=42)
lr.fit(X_train, y_train)
print("Logistic Regression Test Accuracy:", lr.score(X_test, y_test))
```

## 2. Heart Disease Prediction Model
We train a **Support Vector Classifier (SVC)** and compare it with a **Decision Tree Classifier** baseline.

```python
df_heart = pd.read_csv('../data/heart.csv')
X = df_heart.drop(columns=['target'])
y = df_heart['target']

scaler_heart = StandardScaler()
X_scaled = scaler_heart.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

svm = SVC(probability=True, kernel='rbf', C=1.0, random_state=42)
svm.fit(X_train, y_train)
print("SVM Test Accuracy:", svm.score(X_test, y_test))

dt = DecisionTreeClassifier(max_depth=5, random_state=42)
dt.fit(X_train, y_train)
print("Decision Tree Test Accuracy:", dt.score(X_test, y_test))
```

## 3. Chronic Kidney Disease Prediction Model
We train a **K-Nearest Neighbors Classifier** and compare it with a **Gaussian Naive Bayes** baseline.

```python
df_kidney = pd.read_csv('../data/kidney.csv')
X = df_kidney.drop(columns=['class'])
y = df_kidney['class']

scaler_kidney = StandardScaler()
X_scaled = scaler_kidney.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
print("KNN Test Accuracy:", knn.score(X_test, y_test))

nb = GaussianNB()
nb.fit(X_train, y_train)
print("Naive Bayes Test Accuracy:", nb.score(X_test, y_test))
```

## 4. Liver Disease Prediction Model
We train an **XGBoost Classifier** and compare it with a **Gradient Boosting Classifier** baseline.

```python
df_liver = pd.read_csv('../data/liver.csv')
X = df_liver.drop(columns=['Dataset'])
y = df_liver['Dataset']

scaler_liver = StandardScaler()
X_scaled = scaler_liver.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

from xgboost import XGBClassifier
from sklearn.ensemble import GradientBoostingClassifier

xgb = XGBClassifier(eval_metric='logloss', max_depth=4, learning_rate=0.1, random_state=42)
xgb.fit(X_train, y_train)
print("XGBoost Test Accuracy:", xgb.score(X_test, y_test))

gb = GradientBoostingClassifier(max_depth=4, learning_rate=0.1, random_state=42)
gb.fit(X_train, y_train)
print("Gradient Boosting Test Accuracy:", gb.score(X_test, y_test))
```
